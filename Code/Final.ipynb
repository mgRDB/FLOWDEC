```python
def get_embedding(text):
    with torch.no_grad():
        tokens = tokenizer(text, truncation=True, return_tensors='pt').to(0)
        output = model(**tokens)
        embedding = output.last_hidden_state.mean(dim=1).cpu().detach().numpy()
        # embedding = output.pooler_output.detach().numpy()
        # embedding = output.last_hidden_state[:, 0, :].detach().numpy()
    return embedding
```
